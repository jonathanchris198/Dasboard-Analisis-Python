```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
data_day = pd.read_csv('day.csv')

# Konversi kolom 'dteday' ke datetime jika belum
data_day['dteday'] = pd.to_datetime(data_day['dteday'])

# Menghitung matriks korelasi
corr_matrix = data_day.corr()

# Mengatur threshold untuk korelasi yang sangat signifikan
threshold = 0.75  # Anda bisa menyesuaikan nilai ini sesuai kebutuhan

# Membuat mask untuk menampilkan hanya korelasi yang signifikan
mask = (corr_matrix.abs() >= threshold) & (corr_matrix.abs() != 1)

# Membuat heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix[mask], annot=True, cmap='coolwarm', center=0, linewidths=.5)
plt.title('Highest Correlation Daily Dataset')
plt.show()
```

```python
# Load data
data_day = pd.read_csv('day.csv')

# menyesuaikan 'season' sesuai notepad
seasons = {1: 'Springer', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
data_day['season'] = data_day['season'].map(seasons)

# Konversi 'dteday' ke tipe datetime
data_day['dteday'] = pd.to_datetime(data_day['dteday'])
# Membuat kolom 'year' dari kolom 'dteday'
data_day['year'] = data_day['dteday'].dt.year

# Mengelompokkan data berdasarkan musim dan tahun, lalu menghitung total penyewaan
seasonal_data = data_day.groupby(['year', 'season'])['cnt'].sum().unstack()

# Membuat grafik perbandingkan 2011 dan 2012
plt.figure(figsize=(10, 6))
bar_plot = seasonal_data.plot(kind='bar', stacked=False, figsize=(10, 6),
                              colormap='viridis', width=0.7)

# Menambahkan detail pada grafik
plt.title('Comparison of Total Bike Rentals Per Season between 2011 and 2012', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Total Rent', fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title="Season", fontsize=12, loc='upper left')

# Menambahkan label pada setiap bar
for p in bar_plot.patches:
    bar_plot.annotate(format(p.get_height(), ',.0f'),
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha = 'center', va = 'center',
                      xytext = (0, 8), textcoords = 'offset points',
                      fontsize=10, color='black')
# Menampilkan
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

```

- didapatkan jumlah perbandingan di tiap musim antara tahun 2011 dengan tahun 2012.Hal ini berguna dalam membuat event atau promosi di musim yang berpotensi meningkatkan jumlah penyewa sepeda

```python
import pandas as pd
import datetime as dt

# Load dataset
data_day = pd.read_csv('day.csv')

# Pastikan kolom 'dteday' dikonversi ke datetime
data_day['dteday'] = pd.to_datetime(data_day['dteday'])

# Set tanggal analisis sebagai tanggal terbaru di dataset
analysis_date = data_day['dteday'].max()

# 1. RECENCY: Menghitung berapa hari sejak penyewaan terakhir
rfm_recency = data_day.groupby('registered').agg({
    'dteday': lambda x: (analysis_date - x.max()).days
}).reset_index()

# Mengganti nama kolom
rfm_recency.columns = ['registered', 'Recency']

# 2. FREQUENCY: Menghitung jumlah total transaksi (penyewaan) oleh pengguna
rfm_frequency = data_day.groupby('registered').agg({
    'cnt': 'count'  # Jumlah hari pengguna menyewa
}).reset_index()

# Mengganti nama kolom
rfm_frequency.columns = ['registered', 'Frequency']

# 3. MONETARY: Menghitung total penyewaan sepeda oleh pengguna
rfm_monetary = data_day.groupby('registered').agg({
    'cnt': 'sum'  # Total penyewaan yang dilakukan oleh pengguna
}).reset_index()

# Mengganti nama kolom
rfm_monetary.columns = ['registered', 'Monetary']

# Menggabungkan ketiga komponen RFM menjadi satu dataframe
rfm_data = pd.merge(rfm_recency, rfm_frequency, on='registered')
rfm_data = pd.merge(rfm_data, rfm_monetary, on='registered')

# Menampilkan hasil akhir RFM
print(rfm_data.head())
```

## Analisis Lanjutan (Opsional)
- Digunakan Analisis RFM (Recency,Frequency,Monetary) karena cocok dengan data yang ada untuk mengobservasi customer-customer dalam melakukan sewa sepeda yang bermanfaat memaksimalkan profit dalam mengidentifikasi customer yang sering menyewa sepeda dan yang sudah lama tidak menyewa sepeda tetapi frekuensinya melakukan sewa cukup banyak

```python
import datetime as dt

# Ensure that 'dteday' is in datetime format
data_hour['dteday'] = pd.to_datetime(data_hour['dteday'])

# RFM Analisis
# RECENCY
rfm_recency = data_hour.groupby('registered').agg({
    'dteday': lambda x: (analysis_date - x.max()).days
}).reset_index()
rfm_recency.columns = ['registered', 'Recency']

print("Recency calculation:")
print(rfm_recency.head())

# FREQUENCY
rfm_frequency = data_hour.groupby('registered').agg({
    'cnt': 'count'
}).reset_index()
rfm_frequency.columns = ['registered', 'Frequency']

print("\nFrequency calculation:")
print(rfm_frequency.head())

# MONETARY
rfm_monetary = data_hour.groupby('registered').agg({
    'cnt': 'sum'
}).reset_index()
rfm_monetary.columns = ['registered', 'Monetary']

print("\nMonetary calculation:")
print(rfm_monetary.head())

# gabungkan masing-masing R,F,dan M
rfm_data = pd.merge(rfm_recency, rfm_frequency, on='registered')
rfm_data = pd.merge(rfm_data, rfm_monetary, on='registered')

# Menampikan RFM
print("\nFinal RFM Data:")
print(rfm_data.head())
```
